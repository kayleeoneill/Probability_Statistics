```python
for m_guess in np.linspace(-10,10,10):
    for b_guess in np.linspace(-10,10,10):
        
        y_guess = m_guess * x + b_guess

        plt.plot(x,y_guess,'r.');
        plt.plot(x,y,'g.');
        plt.show()

        loss = np.sum((y_guess - y)**2)

        print(loss,m_guess,b_guess)
```

```python
# a double loop with 10 guesses from -10,10 prints loss after each graph
```

```python
best_loss = 100000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,10,20):
    for b_guess in np.linspace(-10,10,20):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess)
```

```python
#a more intelligent search, will keep the best option (puts it in order from worst to best)
```

```python
x = np.linspace(0,10,50) 
m=50
b=100

y=m*x+b
noise = 10 * np.random.randn(50,)
y= m*x+b + noise
plt.plot(x,y,'.')

best_loss = 1000000000000
best_m = 0
best_b = 0 

for m_guess in np.linspace(-10,1000,50):
    for b_guess in np.linspace(-10,1000,50):
        
        y_guess = m_guess * x + b_guess

        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 

            plt.plot(x,y_guess,'r.');
            plt.plot(x,y,'g.');
            plt.show()

            print(loss,m_guess,b_guess)
```
